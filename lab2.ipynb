```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Введенные значения точек
data = np.array([
    [2, 3], [1, 4], [2, 5], [3, 4], [8, 7], [7, 6], [6, 8],
    [8, 3], [6, 2], [9, 5], [4, 8], [5, 9], [1, 7], [0, 8],
    [4, 5], [5, 6], [9, 8], [2, 2], [7, 4], [3, 8], [4, 2],
    [1, 9], [8, 2], [6, 6], [2, 8], [5, 3], [9, 7], [3, 6],
    [1, 2], [7, 8], [3, 9], [8, 6], [4, 7], [6, 4], [2, 7],
    [5, 4], [0, 5], [9, 3], [1, 5], [7, 2], [5, 8], [3, 7]
])

# Количество кластеров
k = 3

# Применение алгоритма k-средних
kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
kmeans.fit(data)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Визуализация результатов кластеризации
colors = ['#FF69B4', '#00CED1', '#0D47A1']

for i in range(k):
    plt.scatter(data[labels == i, 0], data[labels == i, 1], c=colors[i], label=f'Кластер {i+1}', s=30)

plt.scatter(centroids[:, 0], centroids[:, 1], c='k', marker='+', label='Центроид', s=100)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Результаты кластеризации с k-средних')
plt.show()
```
